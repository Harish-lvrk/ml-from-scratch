A lower Gini index indicates a better split. A perfect split (only one class on each side) has a Gini index of 0.

Let's experiment with different depths using a helper function.

````python
def max_depth_error(md):
    model = DecisionTreeClassifier(max_depth=md, random_state=42)
    model.fit(X_train, train_targets)
    train_acc = 1 - model.score(X_train, train_targets)
    val_acc = 1 - model.score(X_val, val_targets)
    return {'Max Depth': md, 'Training Error': train_acc, 'Validation Error': val_acc}
````

````python
%%time
errors_df = pd.DataFrame([max_depth_error(md) for md in range(1, 21)])
````

> **EXERCISE**: Find the combination of `max_depth` and `max_leaf_nodes` that results in the highest validation accuracy.

````python
def test_depth_leaf_nodes(md, mln):
    model = DecisionTreeClassifier(max_depth=md, max_leaf_nodes=mln, random_state=42)
    model.fit(X_train, train_targets)
    train_acc = model.score(X_train, train_targets)
    val_acc = model.score(X_val, val_targets)
    return {'Max Depth': md, 'Max Leaf Nodes': mln, 'Training Accuracy': train_acc, 'Validation Accuracy': val_acc}

depths = range(1, 21)
leaf_nodes = [2**i for i in range(1, 11)] # Testing powers of 2 for leaf nodes

results = []
for md in depths:
    for mln in leaf_nodes:
        results.append(test_depth_leaf_nodes(md, mln))

results_df = pd.DataFrame(results)

# Find the combination with the highest validation accuracy
best_result = results_df.loc[results_df['Validation Accuracy'].idxmax()]

print("Results of hyperparameter tuning:")
display(results_df)

print("\nBest combination of max_depth and max_leaf_nodes:")
display(best_result)
````

## Training a Random Forest

While tuning the hyperparameters of a single decision tree may lead to some improvements, a much more effective strategy is to combine the results of several decision trees trained with slightly different parameters. This is called a random forest.

The key idea here is that each decision tree in the forest will make different kinds of errors, and upon averaging, many of their errors will cancel out. This idea is also known as the "wisdom of the crowd" in common parlance:

<img src="https://i.imgur.com/4Dg0XK4.png" width="480">

A random forest works by averaging/combining the results of several decision trees:

<img src="https://1.bp.blogspot.com/-Ax59WK4DE8w/YK6o9bt_9jI/AAAAAAAAEQA/9KbBf9cdL6kOFkJnU39aUn4m8ydThPenwCLcBGAsYHQ/s0/Random%2BForest%2B03.gif" width="640">


We'll use the `RandomForestClassifier` class from `sklearn.ensemble`.

* Step 1: What is Given and What is Asked

  * Given: The concept of Random Forest.
  * Task: Explain how Random Forest works intuitively (not just technically, but with clear and simple understanding).

>

* Step 2: Intuitive Explanation of Random Forest

  * Start with Decision Trees

    * Imagine you want to predict whether it will rain tomorrow.

    * A single decision tree might ask questions like:

      * Is the humidity high?
      * Is the sky cloudy?
      * Is the wind strong?

    * Based on these questions, it gives you an answer (Yes/No).

  > Problem: One single tree can overfit (be too confident about patterns in the training data).

  * Enter the Forest Idea

    * Instead of relying on just one tree, Random Forest builds many trees.
    * Each tree sees only:

      1. A random sample of the data (not the whole dataset).
      2. A random subset of features (not all input variables).
    * So, every tree is a little bit different.

  * How They Work Together

    * Each tree gives its prediction.
    * The forest then combines these predictions:

      * For classification  takes a majority vote.
      * For regression  takes the average.
    * This way:

      * Mistakes of some trees are corrected by others.
      * The overall result is more stable, less biased, and less overfitted.

  * Analogy: Asking Friends

    * Suppose you want to decide if you should buy a new phone.

    * You ask just one friend  maybe theyre biased (they like iPhones too much ).

    * Instead, you ask 20 friends, each with different experiences.

    * You take the majority opinion.

    * This gives you a more balanced and reliable decision.

  > Thats exactly how Random Forest works  many small opinions combine into one strong, reliable prediction.

* Step 3: Summary of Each Step

  * Step 1: One decision tree makes decisions but can overfit.
  * Step 2: Random Forest grows many trees with randomness (data + features).
  * Step 3: Predictions are aggregated (majority vote or average).

### `min_samples_split` and `min_samples_leaf`

By default, the decision tree classifier tries to split every node that has 2 or more. You can increase the values of these arguments to change this behavior and reduce overfitting, especially for very large datasets.

* **min\_samples\_split**

  > This parameter specifies the minimum number of samples required to split an internal node.
  > By default, it is set to 2, meaning any node with at least 2 samples can be split further.
  > Increasing this value forces the tree to consider only larger nodes for splitting, which reduces model complexity and helps avoid overfitting.

* **min\_samples\_leaf**

  > This parameter specifies the minimum number of samples that a leaf node must have.
  > By default, it is set to 1, meaning a leaf node can contain just a single sample.
  > Increasing this value ensures that leaf nodes have more samples, which makes the model more general and less sensitive to noise in the dataset.

* **Usage Advice**

  > Use **min\_samples\_split** when you want to control how deep the tree can go by restricting splits on small datasets.
  > Use **min\_samples\_leaf** when you want to ensure each leaf is statistically significant and contains enough data points.
  > Both are useful for preventing overfitting, especially when dealing with very large datasets.

* **min\_impurity\_decrease**

  > This parameter is used to control the threshold for splitting nodes in a decision tree.

If the threshold (min\_impurity\_decrease) is set at **0.1**, the split will occur because **0.2 > 0.1**. If the decrease were smaller than 0.1, the split would not occur.
  > By default, it is set to 0, meaning any impurity reduction will allow a split.
  > Increasing this value ensures that only splits with significant impurity reduction are performed, which reduces overfitting and lowers model complexity.

* **Usage Advice**

  > This parameter can be tuned through trial-and-error or guided by educated guesses based on dataset size and complexity.

### `bootstrap`, `max_samples`

By default, a random forest doesn't use the entire dataset for training each decision tree. Instead it applies a technique called bootstrapping. For each tree, rows from the dataset are picked one by one randomly, with replacement i.e. some rows may not show up at all, while some rows may show up multiple times.


<img src="https://i.imgur.com/W8UGaEA.png" width="640">

Bootstrapping helps the random forest generalize better, because each decision tree only sees a fraction of th training set, and some rows randomly get higher weightage than others.

* **bootstrap**

  > In Random Forests, instead of training each decision tree on the full dataset, bootstrapping is applied.
  > Bootstrapping means sampling data **with replacement**. For each tree, rows are picked randomly one by one. Some rows may appear multiple times in the sample, while others may not appear at all.
  > Example: If you have 5 rows `[A, B, C, D, E]`, a bootstrap sample for one tree might be `[B, C, C, E, A]`. Notice that `C` appears twice and `D` is missing.
  > This randomness makes each tree slightly different, helping the forest reduce overfitting and improve generalization.

* **max\_samples**

  > This parameter controls how many rows (samples) are drawn from the dataset to create each bootstrap sample.
  > By default, if `bootstrap=True`, it uses the same number of samples as the original dataset. If the dataset has 1000 rows, each tree also gets 1000 rows (with replacement).
  > If you set `max_samples=0.5`, then each tree will only use 50% of the dataset (randomly sampled with replacement). For 1000 rows, each tree will train on 500 sampled rows.
  > Example: Original dataset = `[A, B, C, D, E]` and `max_samples=3`. One tree may get `[D, A, A]`, another tree may get `[C, E, B]`. Each tree sees a different subset.

* **Usage Advice**

  > Use **bootstrap** to add diversity between trees and prevent them from learning the same patterns.
  > Use **max\_samples** to control how much data each tree sees. Lowering it makes trees more diverse but weaker; higher values make trees stronger but more correlated.
  > The balance of both leads to a strong Random Forest that generalizes well.

* **class\_weight**

  > The `class_weight` parameter is used to handle class imbalance in classification problems.
  > By default, every class is treated equally, meaning the model gives the same importance to all classes, regardless of how frequently they appear in the dataset.
  > When one class is much more frequent than another, the model may become biased toward predicting the majority class. `class_weight` helps solve this by assigning higher weights to minority classes and lower weights to majority classes.

* **How It Works**

  > Suppose you are building a model to detect fraud in transactions:
  >
  > * 95% of the data are **Non-Fraud (Class 0)**
  > * 5% of the data are **Fraud (Class 1)**
  >   Without class weights, the model may just predict everything as **Non-Fraud** and still get 95% accuracy.
  >   If you set `class_weight="balanced"`, the algorithm automatically adjusts weights inversely proportional to class frequencies:
  > * Weight(Class 0) = 1 / 95  0.0105
  > * Weight(Class 1) = 1 / 5 = 0.20
  >   This makes the model pay more attention to Fraud cases.

* **Example**

  > Dataset = `[0, 0, 0, 0, 0, 1]` (5 non-fraud, 1 fraud)
  > Default: Model may learn to always predict `0`.
  > With class weights: The single fraud example will be considered much more important, so the model tries harder to classify it correctly.

* **Usage Advice**

  > Use `class_weight="balanced"` when your dataset has imbalanced classes.
  > You can also manually set weights, e.g., `{0:1, 1:5}` to make fraud detection 5x more important.
  > Proper tuning of class weights improves recall for minority classes while balancing accuracy overall.

#  Sparse vs Dense Explanation

---

##  What is the problem?

* You used an encoder (like `OneHotEncoder`) to transform categorical data.
* The encoder returned a **sparse matrix** instead of a **dense array**.
* When passing this to Pandas, you got an error:

  > `ValueError: Shape of passed values is (1, 1), indices imply (1, 103)`

---

##  Dense Representation

* Dense means **all columns are stored**, even if most are `0`.
* Example: Encoding colors `[Red, Green, Blue]`

  * Input = `"Red"`
  * Dense output = `[1, 0, 0]`
  * Shape = `(1, 3)` (1 row, 3 columns)

> Dense = "Full attendance sheet"  shows everyone, even absentees (zeros).

---

##  Sparse Representation

* Sparse means **only the positions of non-zeros are stored**.
* Example: Encoding the same colors `[Red, Green, Blue]`

  * Input = `"Red"`
  * Sparse output = "Only column 0 is 1"
  * Input = `"Blue"`
  * Sparse output = "Only column 2 is 1"

> Sparse = "Short note"  only lists who is present (non-zeros).

---

##  Why did the Error Happen?

* Your encoder was trained on 103 possible columns.
* Sparse output said: "This row only has **1 active column**."  Shape `(1,1)`
* Pandas expected all **103 columns**  Shape `(1,103)`
* Mismatch caused the error.

> Error happened because Pandas expected the full sheet, but you gave it a short note.

---

##  The Fix

Convert sparse  dense using `.toarray()`:

```python
encoded_array = encoder.transform(input_df[categorical_cols])

if hasattr(encoded_array, "toarray"):   # if its sparse
    encoded_array = encoded_array.toarray()
```

* Now input = `"Red"` becomes `[[1, 0, 0]]` 
* Input = `"Blue"` becomes `[[0, 0, 1]]` 
* Shape = `(1, 3)` matches expected columns.

> Always convert sparse  dense when working with Pandas or DataFrames.

---

##  Final Analogy

* **Dense:** "Attendance register with 100 names  mark 0 or 1 for each."
* **Sparse:** "Just write roll numbers of students who came."
* Your model wanted the **dense register**, but you gave the **short note**.


````python
def predict_input(single_input):
    # Step 1: Convert dictionary input into DataFrame
    input_df = pd.DataFrame([single_input])

    # Step 2: Handle numeric columns
    input_df[numeric_cols] = imputer.transform(input_df[numeric_cols])  # Fill missing values
    input_df[numeric_cols] = scaler.transform(input_df[numeric_cols])  # Apply scaling

    # Step 3: Handle categorical columns (convert sparse -> dense array)
    encoded_array = encoder.transform(input_df[categorical_cols])
    if hasattr(encoded_array, "toarray"):  # if it's sparse
        encoded_array = encoded_array.toarray()

    encoded_df = pd.DataFrame(
        encoded_array,
        columns=encoder.get_feature_names_out(categorical_cols)
    )

    # Step 4: Combine numeric + encoded categorical features
    input_df = input_df.drop(columns=categorical_cols).reset_index(drop=True)
    input_df = pd.concat([input_df, encoded_df], axis=1)

    # Step 5: Ensure correct feature order
    X_input = input_df[numeric_cols + list(encoded_df.columns)]

    # Step 6: Make prediction and probability
    pred = model.predict(X_input)[0]
    prob = model.predict_proba(X_input)[0][list(model.classes_).index(pred)]

    return pred, prob
````
